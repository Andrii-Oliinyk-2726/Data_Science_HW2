```python
# Завдання "Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)"
print("Завдання 'Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)'")
print(80*'_')
missing_percentages = (table.isnull().sum() / len(table)) * 100
print(missing_percentages)
```

```python
# Завдання "Видаліть усі стовпці з пропусками, крім стовпця 'Мова програмування'"
print("Завдання 'Видаліть усі стовпці з пропусками, крім стовпця 'Мова програмування''")
print(80*'_')
# column_to_keep = table.columns[1]
# # table_cleaned = table.dropna(subset=[column_to_keep], axis='columns')
# table_cleaned = table.dropna(axis='columns')

columns_to_keep = ["Язык.программирования"]
for column in table.columns:
    if column != "Язык.программирования":
        if table[column].notnull().all():
            columns_to_keep.append(column)
table_cleaned = table[columns_to_keep]

# print(columns_to_keep)
print(table_cleaned)
```

```python
# Завдання "Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування""
print("Завдання 'Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець 'Мова.програмування''")
print(80*'_')
missing_percentages = (table_cleaned.isnull().sum() / len(table_cleaned)) * 100
print(missing_percentages)
```

```python
# Завдання "Видаліть усі рядки у вихідній таблиці за допомогою методу dropna"
print("Видаліть усі рядки у вихідній таблиці за допомогою методу dropna")
print(80*'_')
table_cleaned_cleaned = table.dropna(axis='columns')
print(table_cleaned_cleaned)
```

```python
# Завдання "Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python"
print("Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python")
print(80*'_')
python_data = table[table["Язык.программирования"] == "Python"]
print(python_data)
```

```python
# Завдання "Використовуючи метод groupby, виконайте групування за стовпчиком "Посада""
print("Використовуючи метод groupby, виконайте групування за стовпчиком 'Посада'")
print(80*'_')
grouped_by_position = table.groupby("Должность")

for position, group_data in grouped_by_position:
    print("Должность:", position)
    print(group_data)
    print("\n")
```

```python
# Завдання "Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць""
print("Створіть новий DataFrame, де для згрупованих даних за стовпчиком 'Посада', виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику 'Зарплата.в.місяць'")
print(80*'_')
grouped_by_position = table.groupby("Должность")

# Виконання агрегації за допомогою методу agg() та знаходження мінімального та максимального значення
aggregated_data = grouped_by_position["Зарплата.в.месяц"].agg(["min", "max"])

print(aggregated_data)
```

```python
# Завдання "Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg""
print("Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик 'avg'")
print(80*'_')
# Функція для розрахунку середньої зарплати на місяць
def fill_avg_salary(row):
    return row["Зарплата.в.месяц"]
# Використання функції fill_avg_salary за допомогою методу apply() для створення стовпчика "avg"
table["avg"] = table.apply(fill_avg_salary, axis=1)
# Виведення таблиці з новим стовпчиком "avg"
print(table)
```

```python
# Завдання "Створіть описову статистику за допомогою методу describe для нового стовпчика."
print("Створіть описову статистику за допомогою методу describe для нового стовпчика.")
print(80*'_')
# Описова статистика для нового стовпчика "avg"
description = table["avg"].describe()
# Виведення описової статистики
print(description)
```

**Візуалізація**

```python
import matplotlib.pyplot as plt
avg_salary_by_position = table.groupby('Должность')['Зарплата.в.месяц'].mean()

# Побудова графіка
plt.figure(figsize=(12, 7))
avg_salary_by_position.plot(kind='bar')
bars = avg_salary_by_position.plot(kind='bar', color='green')
plt.title('Середній розмір зарплати за посадами')
plt.xlabel('Посада', fontsize="small", color="black")
plt.ylabel('Середній розмір зарплати')
plt.xticks(rotation=45, ha='right')
# Додавання значень над стовбчиками
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, round(yval, 2), ha='center', va='bottom', color='black', rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()




```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Групування за должностями та знаходження середньої зарплати
avg_salary_by_position = table.groupby('Должность')['Зарплата.в.месяц'].mean().reset_index()

# Встановлення стилю Seaborn
sns.set(style="whitegrid")

# Побудова графіка з зеленими стовбцями
plt.figure(figsize=(12, 8))
sns.barplot(x='Должность', y='Зарплата.в.месяц', data=avg_salary_by_position, color='red')
plt.title('Середній розмір зарплати за должностями')
plt.xlabel('Посада', color='blue')
plt.ylabel('Середній розмір зарплати', color='blue')
plt.xticks(rotation=45, ha='right')

# Додавання значень над стовбцями під 45 градусів
for index, row in avg_salary_by_position.iterrows():
    plt.text(index, row['Зарплата.в.месяц'] + 50, round(row['Зарплата.в.месяц'], 2), ha='center', va='bottom', color='black', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Побудова кругової діаграми розподілу за стовбчиком "Тип.компании" з проміжками між долями
plt.figure(figsize=(8, 8))
sns.set(style="whitegrid")
type_counts = table['Тип.компании'].value_counts()
explode = (0.01, 0.01, 0.05, 0.05, 0.05)  # Проміжки між долями
plt.pie(type_counts, labels=type_counts.index, shadow=True, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Розподіл за типом компанії')
plt.show()
```
