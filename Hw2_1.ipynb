```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# URL сторінки
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"

# tables = pd.read_html(url, attrs={"id": "collapsibleTable4"})
tables = pd.read_html(url)

# Завдання "Вивести перші рядки таблиці за допомогою методу head"
print("Завдання 'Вивести перші рядки таблиці за допомогою методу head'")
print(80*'_')
print(tables[12].head())
```

```python
# Завдання "Визначити кількість рядків та стовпців у датафреймі (атрибут shape)"
print("Завдання 'Визначити кількість рядків та стовпців у датафреймі (атрибут shape)'")
print(80*'_')
table_1950_2019 = tables[12]
print(table_1950_2019)
num_rows, num_columns = table_1950_2019.shape
print(' ')
print("Кількість рядків:", num_rows)
print("Кількість стовпців:", num_columns)
```

```python
# Завдання "Замінити у таблиці значення "—" на значення NaN"
print(' ')
print("Завдання 'Замінити у таблиці значення '—' на значення NaN'")
print(80*'_')
table_1950_2019 = table_1950_2019.replace('—', np.nan)
print(table_1950_2019)
```

```python
# Завдання "Визначити типи всіх стовпців за допомогою dataframe.dtypes"
print(' ')
print("Завдання 'Визначити типи всіх стовпців за допомогою dataframe.dtypes'")
print(80*'_')
column_types = table_1950_2019.dtypes
print(column_types)
```

```python
# Завдання "Замінити типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—""
print(' ')
print("Завдання 'Замінити типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ '—''")
print(80*'_')
non_numeric_columns = table_1950_2019.columns[1:][table_1950_2019.iloc[:, 1:].dtypes == object]
table_1950_2019[non_numeric_columns] = table_1950_2019[non_numeric_columns].applymap(pd.to_numeric, errors='coerce')
print(table_1950_2019)
print(column_types)
```

```python
# Завдання "Порахувати, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)"
print(' ')
print("Завдання 'Порахувати, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)'")
print(80*'_')
missing_percentages = (table_1950_2019.isnull().sum() / len(table_1950_2019))
print(missing_percentages)
```

```python
# Завдання "Видаліть з таблиці дані по всій країні, останній рядок таблиці"
print(' ')
print("Завдання 'Видаліть з таблиці дані по всій країні, останній рядок таблиці")
print(80*'_')
row_index_to_drop = 27
print(table_1950_2019.drop(index=row_index_to_drop, axis=0))
```

```python
# Завдання "Замінити відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)"
print(' ')
print("Завдання 'Замінити відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)")
print(80*'_')
table_filled = table_1950_2019.fillna(table_1950_2019.mean())
print(table_filled)
```

```python
# Завдання "Отримати список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні"
print(' ')
print("Завдання 'Отримати список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні")
print(80*'_')
mean_value = table_1950_2019["2019"].mean()
filtered_table = table_1950_2019[table_1950_2019["2019"] > mean_value]
# print(mean_value)
# print(values_above_mean)
print(filtered_table)
```

```python
# Завдання "У якому регіоні була найвища народжуваність у 2014 році"
print(' ')
print("Завдання 'У якому регіоні була найвища народжуваність у 2014 році")
print(80*'_')
max_value = table_1950_2019["2014"].max()
print(table_1950_2019[table_1950_2019["2014"] == max_value])
```

```python
# Завдання "Побудувати стовпчикову діаграму народжуваності по регіонах у 2019 році"
print(' ')
print("Завдання 'Побудувати стовпчикову діаграму народжуваності по регіонах у 2019 році")
print(80*'_')
# plt.figure(figsize=(10, 6))  # Налаштування розміру діаграми (необов'язково)
plt.bar(table_1950_2019.index, table_1950_2019['2019'])  # Побудова стовпчикової діаграми
plt.xlabel("Рядки")  # Підпис для осі X (рядки)
plt.ylabel("Значення")  # Підпис для осі Y (значення)
plt.title(f"Стовпчикова діаграма для {'2019'}")  # Заголовок діаграми
plt.xticks(table_1950_2019.index, table_1950_2019["Регіон"], rotation='vertical')  # Налаштування підписів на осі X
plt.tight_layout()  # Налаштування відступів для зручності
plt.show()  # Відображення діаграми
```

```python
# Побудова лінійного графіку за стовбчиками "1950", "2000" та "2019"
plt.figure(figsize=(10, 6))
plt.plot(table_1950_2019.index, table_1950_2019['1950'], marker='o', linestyle='-', color='b', label='1950')
plt.plot(table_1950_2019.index, table_1950_2019['2000'], marker='o', linestyle='-', color='g', label='2000')
plt.plot(table_1950_2019.index, table_1950_2019['2019'], marker='o', linestyle='-', color='r', label='2019')
plt.title('Коефіцієнт народжуваності за роками')
plt.xlabel('Регіон')
plt.ylabel('Коефіцієнт народжуваності')
# plt.xticks(rotation=45, ha='right')
plt.xticks()
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Вибір колонок для графіку
selected_columns = ["1950", "1970", "2000", "2019"]
# Побудова графіку swarmplot для вибраних колонок
plt.figure(figsize=(10, 6))
sns.swarmplot(data=table_1950_2019[selected_columns])
plt.title('Графік swarmplot для 1950, 1970, 2000 та 2019 років')
plt.xlabel('Роки')
plt.ylabel('Значення')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
